It should be noted that the data here has some definitions here that will be listed to ensure that all are aware of what exactly the data above means and is referenceing. The list of definitions are as follows:



>   temp = Mean temperature for the day in degrees Fahrenheit to tenths. Missing 
= 9999.9

>   dewp = Mean dew point for the day in degreesm Fahrenheit to tenths. Missing = 9999.9

> slp = Mean sea level pressure for the day in millibars to tenths. Missing = 9999.9 

> visib = Mean visibility for the day in miles to tenths. Missing = 999.9 

> mxspd =Maximum sustained wind speed reported for the day in knots to tenths. Missing = 999.9

> gust = Maximum wind gust reported for the day in knots to tenths. Missing = 999.9

> min = Minimum temperature reported during the day in Fahrenheit to tenths--time of min temp report varies by country and region, so this will sometimes not be the min for the calendar day. Missing = 9999.9 

> prcp = Total precipitation (rain and/or melted snow) reported during the day in inches and hundredths; will usually not end with the midnight observation--i.e., may include latter part of previous day. .00 indicates no measurable precipitation (includes a trace). Missing = 99.99 Note: Many stations do not report '0' on days with no precipitation--therefore, '99.99' will often appear on these days. Also, for example, a station may only report a 6-hour amount for the period during which rain fell. See Flag field for source of data 

> sndp = Snow depth in inches to tenths--last report for the day if reported more thanonce. Missing = 999.9 Note: Most stations do not report '0' ondays with no snow on the ground--therefore, '999.9' will often appear on these days 

> fog = Indicators (1 = yes, 0 = no/not reported) for the occurrence during the day








**methodology**

Data will be examined in a seiries of graphs to determine wether or not the particular weather condition has any noticable effect on the number of collitions. Each data point on the graphs represents a single day, and data was taken from 2012 to 2020.

**Results**

The following three graphs show the mean, Minimum and maximum tempreture Repectivly. 

```r
plot(data$temp, data$NUM_COLLISIONS, main = 'does the tempreture affect collisions?')
```

```r
plot(data$max, data$NUM_COLLISIONS, main = 'does the maximum tempreture affect collisions?')
```

```r
plot(data$min, data$NUM_COLLISIONS, main = 'does the minimum tempreture affect collisions?')
```

The graphs above do not apear to show any extreme trend, and all three are similar in that they do not have any major spikes that can be taken to show an obvoius trend. However, there are still trends that can be extracted from the graphs. 

There is a void in the data at around the lower then 200 collisions mark and between 0 and 20 degrees. after the 20 degree mark it can be seen that there are a significatnt number of datapoints at the 0 to 200 collisions marks, which then trend slightly upwards and the tempreture goes upwards. 

This trend shows that there is infact a link between tempreture and the number of collisions, where there apears to be a sweet spot from around 20 - 40 degrees where collisions are less likley. 

```r
plot(data$visib, data$NUM_COLLISIONS, main = 'does the visibilty affect collisions?')
```

in the above plot it can be seen that there are less datapoints overall at the lower visibility side of the graph, however what datapoints that are there seem to show higher numbers of collisions. meanwhile there are more datapoints at the higher visibilty end however they have far mor data points at the 0-400 range of collisions, showing that despite there probably being more people out the better visibility seems to make collisions less likley.
